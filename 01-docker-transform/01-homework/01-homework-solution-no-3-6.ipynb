## Question 3. Counting short trips

For the trips in November 2025 (lpep_pickup_datetime between '2025-11-01' and '2025-12-01', exclusive of the upper bound), how many trips had a `trip_distance` of less than or equal to 1 mile?

- 7,853
- 8,007
- 8,254
- 8,421

```python
# Filter `trips` in November 2025
nov_trip = trips[(trips["lpep_pickup_datetime"] >= "2025-11-01") & (trips["lpep_pickup_datetime"] < "2025-12-01")]
```

```python
# Filter `trips` to distance <= 1 mile
nov_trip_1_mile = nov_trip[nov_trip["trip_distance"] <= 1]
```

```python
# Calculate the filtered dataset count
len(nov_trip_1_mile)
```

>Answer: 8,007

## Question 4. Longest trip for each day

Which was the pick up day with the longest trip distance? Only consider trips with `trip_distance` less than 100 miles (to exclude data errors).

Use the pick up time for your calculations.

- 2025-11-14
- 2025-11-20
- 2025-11-23
- 2025-11-25

```python
# Filter `trips` to distance less then 100 miles
trip_less_100_miles = trips[trips["trip_distance"] < 100]
```

```python
# Day with the longest trip distance
trip_less_100_miles[["lpep_pickup_datetime","trip_distance"]].sort_values(by='trip_distance', ascending=False).head()
```

>Answer: 2025-11-14

## Question 5. Biggest pickup zone

Which was the pickup zone with the largest `total_amount` (sum of all trips) on November 18th, 2025?

- East Harlem North
